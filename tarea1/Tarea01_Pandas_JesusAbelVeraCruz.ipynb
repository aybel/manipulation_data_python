
**Enunciado:**
En un entorno de memoria restringida, ¿cómo cargarías eficientemente un 
archivo CSV de 100 GB que contiene un registro histórico de transacciones 
diarias, con el fin de calcular la suma total de las ventas globales?

**Respuesta:**
Para cargar un archivo CSV de 100 GB en un entorno con memoria restringida, 
una estrategia es utilizar el método de **lectura por fragmentos 
(chunking)**. Esta técnica consiste en:

#importar la libreria panda

import pandas as pd

def calcular_suma_total(archivo_path,col_ventas, chunk_tam=100000):

    """
    
    Calcula la suma de ventas procesando el archivo en chunks.
    
    Args:
        archivo_path: Ruta al archivo CSV
        col_ventas: Nombre de la columna con las ventas
        chunk_tam: Número de filas por chunk (ajustable según memoria)
    Returns:
    
        Suma total de ventas
    """
    sumaTotal = 0

    #Iteramos sobre los fragmentos del archivo
    for chunk in pd.read_csv(
        archivoPath,
        chunkSize=chunk_tam,
        usecols=[col_ventas],  # Solo cargar columna necesaria
        dtype={col_ventas:'float64'},
        low_memory=false
    ):
        #Sumamos las ventas del fragmento
        sumaTotal += chunk[col_ventas].sum()
    return sumaTotal

```python
#Ejercicio 3 
### Tienes un archivo Excel con 15 hojas distintas. ¿Cuál es la forma más directa
### y optimizada en Pandas para cargar todas las hojas a la vez en un diccionario
### de DataFrames, evitando abrir y leer el archivo múltiples veces?

# Importar librerías necesarias
import pandas as pd
import os
from datetime import datetime

# ============================================================
# Usando pd.read_excel() con sheet_name=None (RECOMENDADO)
# ============================================================

def cargar_todas_hojas_excel(ruta_archivo):
    """
    Carga todas las hojas de un archivo Excel en un diccionario de DataFrames
    
    Args:
        ruta_archivo (str): Ruta al archivo Excel
        
    Returns:
        dict: Diccionario con nombre de hoja como clave y DataFrame como valor
    """
    # Verificar que el archivo existe
    if not os.path.exists(ruta_archivo):
        raise FileNotFoundError(f"El archivo {ruta_archivo} no existe")
    
    # Cargar todas las hojas en un diccionario (MÁS EFICIENTE)
    # sheet_name=None carga TODAS las hojas en un diccionario
    # El archivo se abre UNA SOLA VEZ
    diccionario_hojas = pd.read_excel(
        ruta_archivo,
        sheet_name=None,  # ¡Clave! Carga todas las hojas a la vez
        engine='openpyxl',  # Motor recomendado para archivos .xlsx
        dtype=str,  # Opcional: leer todo como string para evitar inferencias
        na_filter=False  # Opcional: no filtrar NAs para mejorar rendimiento
    )
    
    return diccionario_hojas
```
